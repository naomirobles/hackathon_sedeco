```python
import pandas as pd
import geopandas as gpd

CRS_GEO     = 'EPSG:4326'   # geográfico (almacenamiento)
```

```python

poi_geo  = gpd.read_file("assets/poligonos/POI/capas_POIs_eliminar.gpkg").to_crs(CRS_GEO)
```

```python
# Lista de columnas a eliminar
columnas_a_eliminar = ['path', 'CONCAT_STV 9M Y 20M',
                       'ID_9m', 'ID_CARTO', 'ID_20m',
                       'SUMA 9M', 'SUMA 20M', 'COBERTURA_STV',
                       'COB_TOTEM', 'SUMA TOTEM', 'COBERTURA_TOTEM',
                       'CON COBERTURA_EN GENERAL', 'SIN COBERTURA_EN GENERAL']

# Eliminar las columnas (solo las que existan en el GeoDataFrame)
columnas_existentes = [col for col in columnas_a_eliminar if col in poi_geo.columns]
poi_geo = poi_geo.drop(columns=columnas_existentes)

# Guardar como GeoPackage (sobrescribe el archivo original o crea uno nuevo)
poi_geo.to_file("assets/poligonos/POI/capas_POIs.gpkg", driver='GPKG')
```
